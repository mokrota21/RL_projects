# Full vision update_step = 1

```python
import json
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Function to load JSON file
def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Load JSON files
count = 0
files = []
data = []
filenames = []
while True:
    file = '../vision_' + str(count) + '.json'
    try:
        data.append(load_json(file_path=file))
        filenames.append(file)
        print(file)
    except:
        break
    count += 1

for filepath in files:
    data.append(load_json(file_path=filepath))

# Plot comparison of a specific field using Plotly
def plot_comparison(fields, data, random_last=False):
    fig = make_subplots(rows=1, cols=1, subplot_titles=[f'Comparison of {fields}'])
    
    # Add traces for each file
    for field in fields:
        for i, d in enumerate(data):
            field_data = d.get(field, [])
            if len(field_data) > 0 and isinstance(field_data[0], bool):
                field_data = list(map(lambda x: float(x), field_data))
            if i == len(data) - 1 and random_last:
                fig.add_trace(go.Scatter(y=field_data, mode='lines', name=f'Random rewards'))
            else:
                fig.add_trace(go.Scatter(y=field_data, mode='lines', name=f'{filenames[i]} - {field}'))
    # Customize layout
    fig.update_layout(
        title=f'Comparison of {field}',
        xaxis_title='Step',
        yaxis_title=field,
        showlegend=True,
        hovermode="x"
    )
    
    # Show the interactive plot
    fig.show()
    return fig

# Plot losses
loss_fig = plot_comparison(['loss'], data)


# random = load_json('../random_rewards.json')
# data.append(random)
# Plot avg_rewards
reward_fig = plot_comparison(['reward'], data)
```

```python
# Initialize a list to hold indices of traces to delete
delete_indices = []

# Step 1: Collect indices of traces where last reward < -4
for j, trace in enumerate(reward_fig.data):
        if trace.y[-1] < -4:
            delete_indices.append(j)
```

```python
for fig in [loss_fig, reward_fig]:
    fig.data = tuple(trace for i, trace in enumerate(fig.data) if i not in set(delete_indices))
```

```python
loss_fig.show()
reward_fig.show()
```

```python
import json
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Function to load JSON file
def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Load JSON files
count = 1
files = []
data = []
filenames = []
while True:
    file = '../full_vision_long_run_' + str(count) + '.json'
    try:
        data.append(load_json(file_path=file))
        filenames.append(file)
        print(file)
    except:
        break
    count += 1
```

```python
res = [{}]

for k in data[0].keys():
    total = []
    for d in data:
        total += d[k]
    res[0][k] = total
```

```python
loss_fig = plot_comparison(['loss'], res)

reward_fig = plot_comparison(['reward'], res)
```
