```python
import json
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Function to load JSON file
def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Load JSON files
files = ['../metrics_data_' + str(i) + '.json' for i in range(1, 30)]

data = []

for filepath in files:
    data.append(load_json(file_path=filepath))

# Plot comparison of a specific field using Plotly
def plot_comparison(field, data, random_last=False):
    fig = make_subplots(rows=1, cols=1, subplot_titles=[f'Comparison of {field}'])
    
    # Add traces for each file
    for i, d in enumerate(data):
        field_data = d.get(field, [])
        if i == len(data) - 1 and random_last:
            fig.add_trace(go.Scatter(y=field_data, mode='lines', name=f'Random rewards'))
        else:
            fig.add_trace(go.Scatter(y=field_data, mode='lines', name=f'File {i+1} - {field}'))
    # Customize layout
    fig.update_layout(
        title=f'Comparison of {field}',
        xaxis_title='Step',
        yaxis_title=field,
        showlegend=True,
        hovermode="x"
    )
    
    # Show the interactive plot
    fig.show()

# Plot losses
plot_comparison('losses', data)


random = load_json('../random_rewards.json')
data.append(random)
# Plot avg_rewards
plot_comparison('avg_rewards', data, random_last=True)

```
