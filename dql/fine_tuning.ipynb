```python
import json
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Function to load JSON file
def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Load JSON files
count = 1
files = []
data = []
while True:
    file = '../nlearning_' + str(count) + '.json'
    try:
        data.append(load_json(file_path=file))
        print(file)
    except:
        break
    count += 1


for filepath in files:
    data.append(load_json(file_path=filepath))

# Plot comparison of a specific field using Plotly
def plot_comparison(fields, data, random_last=False):
    fig = make_subplots(rows=1, cols=1, subplot_titles=[f'Comparison of {fields}'])
    
    # Add traces for each file
    for field in fields:
        for i, d in enumerate(data):
            field_data = d.get(field, [])
            if len(field_data) > 0 and isinstance(field_data[0], bool):
                field_data = list(map(lambda x: float(x), field_data))
            if i == len(data) - 1 and random_last:
                fig.add_trace(go.Scatter(y=field_data, mode='lines', name=f'Random rewards'))
            else:
                fig.add_trace(go.Scatter(y=field_data, mode='lines', name=f'File {i+1} - {field}'))
    # Customize layout
    fig.update_layout(
        title=f'Comparison of {field}',
        xaxis_title='Step',
        yaxis_title=field,
        showlegend=True,
        hovermode="x"
    )
    
    # Show the interactive plot
    fig.show()

# Plot losses
plot_comparison(['loss', 'goal', 'subgoal'], data)


random = load_json('../random_rewards.json')
data.append(random)
# Plot avg_rewards
plot_comparison(['reward'], data, random_last=True)

```
